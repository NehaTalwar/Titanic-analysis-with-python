#  TITANIC ANALYSIS


In this analysis i will analyze the Titanic data using Python ,Pandas ,Matplotlib, Numpy to create attractive Visualizations that will inform us about the survival dynamics among the passengers based on Gender , Age , Class who happened to be on the infamous Titanic on it's maiden voyage. There is a sharp skew in favour of the female passengers who were on-board compared to the male passengers and the data reflects the same. 





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("Titanic2.csv", squeeze = True)
titanic.head(10)
```

```python
titanics = titanic [["Sex" , "SurvivedCategory"]]
titanics.head(10)
```

```python
females = titanics.loc[titanics["Sex"] == "female"]
```

```python
survived_f = females.loc[females["SurvivedCategory"]=="Survived"]
```

''' GENDERWISE SURVIVAL AND DEATH RATE OF TITANIC PASSENGERS'''

```python

xaxis =  ['Survived Females' , 'Died females' , 'Survived Males' , 'Died Males']
tick_locations = [x for x in xaxis]


number = [233  , 81 ,  109 , 468 ]


plt.figure(figsize= (20,10))

plt.bar( xaxis, number , alpha = 0.75, align = "center", color=[ 'hotpink', 'cyan', 'blue', 'red'] )

plt.title("GENDERWISE SURVIVAL STATISTICS ON THE\n TITANIC" , size = 30)
plt.xlabel("Genderwise Survival and Death" , size = 25 , color = "red")
plt.ylabel("Number of Passengers" , size= 25 , color = "red")

plt.xlim(-0.75,len(xaxis)-0.25)
plt.ylim(0, max(number) + 35)
plt.yticks(size = 18 , color ="brown")
plt.xticks( tick_locations, xaxis , rotation= 45 , size = 18 , color = "brown")

plt.show()


```



A total of 891 passengers were on board of which 314 were females , a total of 233 females survived and 81 died which means almost 26% of female passengers were survivors of a total of 891 on board , the "did not survive" female passengers represent 9% of the total pax. The total no of male passengers on board were 577 of which 109 survived the catastrophe and 468 did not survive ,the percentages of survived and 'did not survive' male pax are 12% and 52% respectively. Keeping the figures in mind it would be fair to say that gender influenced survival rates to a huge degree and "Ladies and children first" was definetely a huge factor while boarding the lifeboats to rescue passengers.



 ''' AGEWISE SURVIVAL OF FEMALES AND MALES '''

```python

titanica = titanic [[ "SurvivedCategory", "Sex" , "Age" ]]

females = titanica.loc[titanica["Sex"] == "female"]
survived_f = females.loc[females["SurvivedCategory"]=="Survived"]
f_age = survived_f[["Age"]]
clean_f_age = f_age.dropna(how = "any")
clean_f_age.head()
```

```python
clean_f_age = np.array(clean_f_age)
```

```python

group1 =[]
group2 =[]
group3 =[]

for x in clean_f_age:
    if (x <= 25):
        group1.append(x)
    elif (x <= 50):
        group2.append(x)
    else:
        group3.append(x)
         
  
```

```python
len(group3)
```

```python
#group1 = 87 females
#group2 = 94 females
#group3 = 16 females
```

```python

xaxis =  ["age(0 to 25)" , "age(26 to 50)" , "age(51 and above)"]
tick_locations = [x for x in xaxis]


age_groups = [87 ,94 ,16]


plt.figure(figsize= (20,10))

plt.bar( xaxis, age_groups , alpha = 0.75, align = "center", color=[ 'lightcoral', 'magenta', 'mediumvioletred'] )

plt.title("AGEWISE FEMALE SURVIVAL STATISTICS ON THE \nTITANIC" , size = 30)
plt.xlabel("Agewise Survival of Females" , size = 25 , color = "red")
plt.ylabel("Number of Passengers" , size= 25 , color = "red")

plt.xlim(-0.75,len(xaxis)-0.25)
plt.ylim(0, max(age_groups) + 10)

plt.xticks( tick_locations, xaxis , rotation= 45 , size = 18 , color = "crimson")
plt.yticks(size = 18 , color = "crimson")
plt.grid(alpha= 0.25)
plt.show()


```



Of a total of 197 Females who survived and whose ages were known ,87 were in the age group of 0-25 which is 44% , 94 were between the ages of 26 to 50 converting to the highest percentage of survival in this age group which is 47% and 16 were in the age group of 51 and above which is almost 8% . Clearly , the younger aged female had an exponentially higher chance of survival compared to ages 50 and above.

''' CLASSWISE SURVIVAL RATE OF FEMALES AND MALES '''

```python

titanicc = titanic [["SurvivedCategory", "Class" ,"Sex"]]
females = titanicc.loc[titanicc["Sex"] == "female"]
survived_f = females.loc[females["SurvivedCategory"]=="Survived"]
class_f = survived_f["Class"].value_counts()

class_f
```

```python

class_nos = [ 91,72,70 ]
labels = ["First Class" ,"Coach" , "Business Class" ]
colors= [ "lightpink" , "thistle", "Plum"]
explode = (0.10,  0 , 0)

plt.figure(figsize=(30,20))
plt.title("% OF TOTAL FEMALES SURVIVED BY CLASS TYPE" , size =35, color ="m")
plt.pie(class_nos, explode=explode, labels=labels ,colors=colors, autopct="%1.1f%%", shadow=True, startangle=180,
        textprops={'fontsize': 30})
plt.axis("equal")
plt.show

```



The class division of the 233 females passengers who survived the sinking is clearly shown in the pie chart here with a the biggest share of the pie belonging to those 91 first class female passengers at 39% and then from there an almost equal survival ratio with coach class ladies at 30.9% or 72 total followed by Business class at 30% or 70 ladies. This shows the highest chances of survival were indeed with the First class female passengers in the ill-fated ship.



```python

titaniccm = titanic [["SurvivedCategory", "Class" ,"Sex"]]
Males = titaniccm.loc[titaniccm["Sex"] == "male"]
survived_m = Males.loc[Males["SurvivedCategory"]=="Survived"]
class_m = survived_m["Class"].value_counts()

class_m
```

```python

class_nos = [ 45 , 47 ,17 ]
labels = ["First Class" ,"Coach" , "Business Class" ]
colors= [ "mediumspringgreen" , "darkturquoise", "mediumslateblue"]
explode = (0.10,  0 , 0)

plt.figure(figsize=(30,20))
plt.title("% OF TOTAL MALES SURVIVED BY CLASS TYPE" , size =35 , color ="g")
plt.pie(class_nos, explode=explode, labels=labels ,colors=colors, autopct="%1.1f%%", shadow=True, startangle=150,
        textprops={'fontsize': 30})
plt.axis("equal")
plt.show

```



The males who survived had an intriguing passenger class profile with the Coach class having the highest percentage of survivors at 43% or 47 pax , the first class male passengers are at 41% or 45 total and the business class passengers at 15% or 17 total.It reflects that class did not set precedence instead the coach class pax survived despite the class difference from first and business class male passengers unlike the female passengers.
